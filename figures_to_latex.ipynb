```python
import glob
import os
```

```python
def my_model_graphics(name):
    caption = lambda x: f"Test performance of the models developed in this thesis on recording {x}. The plots include the target labels, model output probabilities, predictions and performance metrics"
    label = lambda x: "\label\{fig\:my_models_%s\}" % x
    png = lambda x: f"{x}.png"
    graphic_command = lambda x: f"width={x}\\textwidth"

    out = """
    \\begin{figure}
        \centering
        \includegraphics[%s]{appendices/Results/img/my_models/%s}
        \caption{%s}
        \label{fig:my_models_%s}
    \end{figure}""" % (graphic_command(1), png(name),
                           caption(name),
                           name)
                        
    return out

my_model_files = glob.glob("Matlab/new_main_comparison/figures/my_models/*.png")
my_model_files = [f.split("\\")[-1].replace(".png","") for f in my_model_files]

for f in my_model_files:
    print(my_model_graphics(f))
```

```python
def all_model_graphics(name):
    caption = lambda x: f"Test performance of the models developed in this thesis and benchmark models on recording {x}. The plots include the target labels, model output predictions and the corresponding performance metrics"
    label = lambda x: "\label\{fig\:all_models_%s\}" % x
    png = lambda x: f"{x}.png"
    graphic_command = lambda x: f"width={x}\\textwidth"

    out = """
    \\begin{figure}
        \centering
        \includegraphics[%s]{appendices/Results/img/all_models/%s}
        \caption{%s}
        \label{fig:all_models_%s}
    \end{figure}""" % (graphic_command(1), png(name),
                           caption(name),
                           name)
                        
    return out

all_model_files = glob.glob("Matlab/new_main_comparison/figures/all_models/*.png")
all_model_files = [f.split("\\")[-1].replace(".png","") for f in all_model_files]

for f in my_model_files:
    print(all_model_graphics(f))
```

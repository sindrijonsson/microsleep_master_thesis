```python
import numpy as np
import scipy.io
from glob import glob
import os
```

```python
npy_files = glob("predictions/*/*.npy")
npy_files
```

```python
for file in npy_files:
    tmp = np.load(file)
    out_file = f"Matlab\\{file.replace('.npy','.mat')}"
    scipy.io.savemat(out_file, dict(data=tmp))
```
